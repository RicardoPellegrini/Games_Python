Neste jogo da velha, podemos jogar contra o computador. Ele vai fazer escolhas lógicas, tentando sempre finalizar o jogo e também impedindo você de ganhar.

```python
# Importando bibliotecas
from random import choice
from time import sleep
```

```python
# Mostrando tabuleiro
def mostra_tabuleiro():
  global pos
  print(f' {pos[0]} | {pos[1]} | {pos[2]} ')
  print(f' {pos[3]} | {pos[4]} | {pos[5]} ')
  print(f' {pos[6]} | {pos[7]} | {pos[8]} ')
```

```python
# Escolhendo simbolo
def define_simbolo():
  global simbolo_j, simbolo_comp
  simbolo_j = input('Escolha seu simbolo [X/O]: ').upper()[0]
  while True:
      if simbolo_j in ['X','O']:
        print(f'Seu simbolo é o {simbolo_j}')
        break
      else:
        simbolo_j = input('Escolha seu simbolo entre [X/O]: ').upper()[0]
  if simbolo_j == 'X':
    simbolo_comp = 'O'
  else:
    simbolo_comp = 'X'
```

```python
# Sorteando quem comeca
def define_prim_jogador():
  global prim_jogador, vez
  prim_jogador = choice([1,2])
  if prim_jogador == 1:
    print('Você começa!')
    vez = 'Jogador'
  else:
    print('O computador começa!')
    vez = 'Computador'
  return prim_jogador
```

```python
# Escolhendo jogada
def jogada():
  global pos, simbolo_j, simbolo_comp, vez
  if vez == 'Jogador':
    while True:
      try:
        jog = int(input('Escolha uma posicao entre 1 e 9: '))
        if jog in range(1,10) and pos[jog-1] == '-':
          pos[jog-1] = simbolo_j
          break
      except:
        print('Você deve inserir um valor disponivel no tabuleiro: ')
  else:
    print('O computador está escolhendo uma posicao: ')
    sleep(3)
    jogada_computador()
    

# Algoritmo para definir jogada do computador mediante jogadas passadas
def jogada_computador():
  global pos, simbolo_comp, simbolo_j
  # Primeira linha
  if (pos[0] == pos[1] == simbolo_comp and pos[2] == '-'):
    pos[2] = simbolo_comp
  elif (pos[0] == pos[2] == simbolo_comp and pos[1] == '-'):
    pos[1] = simbolo_comp
  elif (pos[1] == pos[2] == simbolo_comp and pos[0] == '-'):
    pos[0] = simbolo_comp
  # Segunda linha
  elif (pos[3] == pos[4] == simbolo_comp and pos[5] == '-'):
    pos[5] = simbolo_comp
  elif (pos[3] == pos[5] == simbolo_comp and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[5] == simbolo_comp and pos[3] == '-'):
    pos[3] = simbolo_comp
    # Terceira linha
  elif (pos[6] == pos[7] == simbolo_comp and pos[8] == '-'):
    pos[8] = simbolo_comp
  elif (pos[6] == pos[8] == simbolo_comp and pos[7] == '-'):
    pos[7] = simbolo_comp
  elif (pos[7] == pos[8] == simbolo_comp and pos[6] == '-'):
    pos[6] = simbolo_comp
    # Primeira coluna
  elif (pos[0] == pos[3] == simbolo_comp and pos[6] == '-'):
    pos[6] = simbolo_comp
  elif (pos[0] == pos[6] == simbolo_comp and pos[3] == '-'):
    pos[3] = simbolo_comp
  elif (pos[3] == pos[6] == simbolo_comp and pos[0] == '-'):
    pos[0] = simbolo_comp
    # Segunda coluna
  elif (pos[1] == pos[4] == simbolo_comp and pos[7] == '-'):
    pos[7] = simbolo_comp
  elif (pos[1] == pos[7] == simbolo_comp and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[7] == simbolo_comp and pos[1] == '-'):
    pos[1] = simbolo_comp
    # Teceira coluna
  elif (pos[2] == pos[5] == simbolo_comp and pos[8] == '-'):
    pos[8] = simbolo_comp
  elif (pos[2] == pos[8] == simbolo_comp and pos[5] == '-'):
    pos[5] = simbolo_comp
  elif (pos[5] == pos[8] == simbolo_comp and pos[2] == '-'):
    pos[2] = simbolo_comp
    # Diagonal principal
  elif (pos[0] == pos[4] == simbolo_comp and pos[8] == '-'):
    pos[8] = simbolo_comp
  elif (pos[0] == pos[8] == simbolo_comp and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[8] == simbolo_comp and pos[0] == '-'):
    pos[0] = simbolo_comp
    # Diagonal secundaria
  elif (pos[6] == pos[4] == simbolo_comp and pos[2] == '-'):
    pos[2] = simbolo_comp
  elif (pos[6] == pos[2] == simbolo_comp and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[2] == simbolo_comp and pos[6] == '-'):
    pos[6] = simbolo_comp
  # Primeira linha considerando jogada do adversario
  elif (pos[0] == pos[1] == simbolo_j and pos[2] == '-'):
    pos[2] = simbolo_comp
  elif (pos[0] == pos[2] == simbolo_j and pos[1] == '-'):
    pos[1] = simbolo_comp
  elif (pos[1] == pos[2] == simbolo_j and pos[0] == '-'):
    pos[0] = simbolo_comp
  # Segunda linha considerando jogada do adversario
  elif (pos[3] == pos[4] == simbolo_j and pos[5] == '-'):
    pos[5] = simbolo_comp
  elif (pos[3] == pos[5] == simbolo_j and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[5] == simbolo_j and pos[3] == '-'):
    pos[3] = simbolo_comp
    # Terceira linha considerando jogada do adversario
  elif (pos[6] == pos[7] == simbolo_j and pos[8] == '-'):
    pos[8] = simbolo_comp
  elif (pos[6] == pos[8] == simbolo_j and pos[7] == '-'):
    pos[7] = simbolo_comp
  elif (pos[7] == pos[8] == simbolo_j and pos[6] == '-'):
    pos[6] = simbolo_comp
    # Primeira coluna considerando jogada do adversario
  elif (pos[0] == pos[3] == simbolo_j and pos[6] == '-'):
    pos[6] = simbolo_comp
  elif (pos[0] == pos[6] == simbolo_j and pos[3] == '-'):
    pos[3] = simbolo_comp
  elif (pos[3] == pos[6] == simbolo_j and pos[0] == '-'):
    pos[0] = simbolo_comp
    # Segunda coluna considerando jogada do adversario
  elif (pos[1] == pos[4] == simbolo_j and pos[7] == '-'):
    pos[7] = simbolo_comp
  elif (pos[1] == pos[7] == simbolo_j and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[7] == simbolo_j and pos[1] == '-'):
    pos[1] = simbolo_comp
    # Teceira coluna considerando jogada do adversario
  elif (pos[2] == pos[5] == simbolo_j and pos[8] == '-'):
    pos[8] = simbolo_comp
  elif (pos[2] == pos[8] == simbolo_j and pos[5] == '-'):
    pos[5] = simbolo_comp
  elif (pos[5] == pos[8] == simbolo_j and pos[2] == '-'):
    pos[2] = simbolo_comp
    # Diagonal principal considerando jogada do adversario
  elif (pos[0] == pos[4] == simbolo_j and pos[8] == '-'):
    pos[8] = simbolo_comp
  elif (pos[0] == pos[8] == simbolo_j and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[8] == simbolo_j and pos[0] == '-'):
    pos[0] = simbolo_comp
    # Diagonal secundaria considerando jogada do adversario
  elif (pos[6] == pos[4] == simbolo_j and pos[2] == '-'):
    pos[2] = simbolo_comp
  elif (pos[6] == pos[2] == simbolo_j and pos[4] == '-'):
    pos[4] = simbolo_comp
  elif (pos[4] == pos[2] == simbolo_j and pos[6] == '-'):
    pos[6] = simbolo_comp
  else:
    jog = choice(range(9))
    while True:
      if pos[jog] == '-':
        pos[jog] = simbolo_comp
        break
      else:
        jog = choice(range(9))
    return
```

```python
# Alterando vez do jogador
def alterar_jogador():
  global vez, simbolo_j, simbolo_comp, resultado
  if vez == 'Jogador':
    vez = 'Computador'
    print(f'Agora é a vez do computador ({simbolo_comp})')
  else:
    vez = 'Jogador'
    print(f'Agora é a sua vez ({simbolo_j})')
  return vez
```

```python
# Verificando se houve vitoria ou empate
def verifica_vitoria():
  global pos, vez, simbolo_comp, simbolo_j, simbolo, resultado
  if vez == 'Jogador':
    simbolo = simbolo_j
  else:
    simbolo = simbolo_comp

  if (pos[0] == pos[1] == pos[2] == simbolo) or\
     (pos[3] == pos[4] == pos[5] == simbolo) or\
     (pos[6] == pos[7] == pos[8] == simbolo):
      print(f'Vitoria do {vez}, de simbolo {simbolo}!')
      resultado = 'Vitoria'
      return fim_de_jogo()
  elif (pos[0] == pos[3] == pos[6] == simbolo) or\
      (pos[1] == pos[4] == pos[7] == simbolo) or\
      (pos[2] == pos[5] == pos[8] == simbolo):
      print(f'Vitoria do {vez}, de simbolo {simbolo}!')
      resultado = 'Vitoria'
      return fim_de_jogo()
  elif (pos[0] == pos[4] == pos[8] == simbolo) or\
      (pos[6] == pos[4] == pos[2] == simbolo):
      print(f'Vitoria do {vez}, de simbolo {simbolo}!')
      resultado = 'Vitoria'
      return fim_de_jogo()
  else:
    if pos.count('-') == 0:
      print('Empate!')
      resultado = 'Empate'
      return fim_de_jogo()
    else:
      return alterar_jogador()
```

```python
def fim_de_jogo():
    global resultado, simbolo, vez, novamente
    if resultado == 'Vitoria' or resultado == 'Empate':
        print('-' * 12)
        print('Fim de jogo!')
        print('-' * 12)
    novamente = input('Deseja jogar novamente? [S/N]: ').upper()[0]
    while True:
        if novamente not in ['S','N']:
            novamente = input('Deseja jogar novamente? [S/N]: ').upper()[0]
        else:
            break
```

```python
# Estrutura final do jogo
def jogar():
    global pos, resultado, novamente
    novamente = None
    while novamente != 'N':
        print('=' * 32)
        print('SEJA BEM-VINDO AO JOGO DA VELHA!')
        print('=' * 32)
        # Tabuleiro inicial
        pos = ['-' for i in range(9)]
        mostra_tabuleiro()
        define_simbolo()
        define_prim_jogador()
        resultado = None
        while resultado not in ['Vitoria', 'Empate']:
            jogada()
            mostra_tabuleiro()
            verifica_vitoria()
        if novamente == 'N':
            print('=' * 33)
            print('Obrigado pelo jogo, até a próxima!')
            print('=' * 33)
            break
```

```python
if __name__ == '__main__':
    jogar()
```
